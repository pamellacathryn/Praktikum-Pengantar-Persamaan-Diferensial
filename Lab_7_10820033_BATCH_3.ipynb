```python
from sympy import *
from sympy.abc import t
from sympy.utilities.lambdify import lambdify
import numpy as np
import matplotlib.pyplot as plt

'''
f(x)=x-x^3, -1<x<1
n = 5, 10, 20
'''

def f(x):
  return x-x**3
s = fourier_series(f(t), (t, -1, 1))
N = [5,10,20] #banyak k suku pertama yang dipertimbangkan
for k in N:
  fun_fourier = lambdify(t, s.truncate(k))
  x = np.linspace(-1, 1, 100)
  plt.plot(x, f(x), linewidth = 2)
  plt.plot(x, fun_fourier(x), linewidth = 2)
  pam = np.linspace(-3, -1, 100)
  plt.plot(pam, f(pam+2), linewidth = 2)
  caca = np.linspace(1, 3, 100)
  plt.plot(caca, f(caca-2), linewidth = 2)
  agus = np.linspace(3, 5, 100)
  plt.plot(agus, f(agus-4), linewidth = 2)
  plt.grid()
  plt.legend([f'f(t)={f(t)}',f'Deret Fourier ({k} suku pertama)'])
  plt.show()
  print()
```

```python
from sympy import *
from sympy.abc import t
from sympy.utilities.lambdify import lambdify
import numpy as np
import matplotlib.pyplot as plt
#PERLUASAN GANJIL

#a.
print("a. Perluasan Ganjil")
s = fourier_series(sin(t), (t, -2*pi, 2*pi)) #f(t), tE[0,pi]
v = fourier_series(-sin(-t), (t, -2*pi, 2*pi)) #f(-t), tE[-pi,0]
print(s)
print()

#b.
print("b.")
N = [3,5,10] #banyak k suku pertama yang dipertimbangkan
for k in N:
  fun_fourier_1 = lambdify(t, s.truncate(k))
  fun_fourier_2 = lambdify(t, v.truncate(k))
  x = np.linspace(0, 2*np.pi, 100)
  plt.plot(x, np.sin(x), linewidth = 2)
  y = np.linspace(-2*np.pi, 0, 100)
  plt.plot(y, -np.sin(-x), linewidth = 2)
  plt.plot(x, fun_fourier_1(x), linewidth = 2)
  plt.plot(y, fun_fourier_2(y), linewidth = 2)
  plt.grid()
  plt.legend([f'f(t)={f(t)}',f'f(t)={-f(-t)}'])
  plt.xlabel("t")
  plt.ylabel("f(t)")
  plt.title(f"Deret Fourier ({k} suku pertama)")
  plt.show()
print()

#PERLUASAN GENAP
#c.
print("c. Perluasan Genap")
s = fourier_series(sin(t), (t, -2*pi, 2*pi)) #f(t), tE[0,pi]
v = fourier_series(sin(-t), (t, -2*pi, 2*pi)) #f(-t), tE[-pi,0]
print(s)
print()

#d.
print("d.")
N = [3,5,10] #banyak k suku pertama yang dipertimbangkan
for k in N:
  fun_fourier_1 = lambdify(t, s.truncate(k))
  fun_fourier_2 = lambdify(t, v.truncate(k))
  x = np.linspace(0, 2*np.pi, 100)
  plt.plot(x, np.sin(x), linewidth = 2)
  y = np.linspace(-2*np.pi, 0, 100)
  plt.plot(y, np.sin(-x), linewidth = 2)
  plt.plot(x, fun_fourier_1(x), linewidth = 2)
  plt.plot(y, fun_fourier_2(y), linewidth = 2)
  plt.grid()
  plt.legend([f'f(t)={f(t)}',f'f(t)={-f(-t)}'])
  plt.xlabel("t")
  plt.ylabel("f(t)")
  plt.title(f"Deret Fourier ({k} suku pertama)")
  plt.show()
print()
```
