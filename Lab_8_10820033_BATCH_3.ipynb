```python
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
from matplotlib import cm
from matplotlib.ticker import LinearLocator, FormatStrFormatter
from sympy import *
import scipy.integrate as integrate
import numpy as np

'''
Persamaan Laplace
u_t = u_xx
0<x<1
u(0,t)=10
u(1,t)=20
u(x,0)=1
'''

fig = plt.figure()
ax = fig.gca(projection="3d")

#Domain Pengamatan
x = np.linspace(0,1,250)
t = np.linspace(0,500,250)
x,t = np.meshgrid(x,t)

#Parameter
u0 = 1 #u(x,0)
k = 1 #u_t = k*u_xx
L = 1 #panjang batang

#a.
print('''a.
diperoleh solusi formal (perhitungan cn di cell sebelumnya)
u(x,t)=10x+10+ sigma dari n=1 sampai inf [ (38 cos(npi)-18)/npi * e^(-n^2 pi^2 t) * sin(npix) ]
''')

#b.
print("b. untuk N=3")
#Reset nilai U
U = 0

#Melakukan perhitungan estimasi nilai u(x,t)
len_sim = 3 #Hampiran solusi (len_sim suku pertama)
for n in range(1,len_sim):
  U = U + ((38*np.cos(n*np.pi)-18)/(n*np.pi))*np.exp(-n**2*np.pi**2*t)*np.sin(n*np.pi*x)
U = 10*x+10+U

#Visualisasi
surf = ax.plot_surface(x,t,U,
                       cmap=cm.coolwarm,
                       linewidth=0,
                       antialiased=False)
fig.colorbar(surf,
             shrink=0.5,
             aspect=5)
plt.show()
print()

print("untuk N=5")
fig = plt.figure()
ax = fig.gca(projection="3d")

#Reset nilai U
U = 0

#Melakukan perhitungan estimasi nilai u(x,t)
len_sim = 5 #Hampiran solusi (len_sim suku pertama)
for n in range(1,len_sim):
  U = U + ((38*np.cos(n*np.pi)-18)/(n*np.pi))*np.exp(-n**2*np.pi**2*t)*np.sin(n*np.pi*x)
U = 10*x+10+U

#Visualisasi
surf = ax.plot_surface(x,t,U,
                       cmap=cm.coolwarm,
                       linewidth=0,
                       antialiased=False)
fig.colorbar(surf,
             shrink=0.5,
             aspect=5)
plt.show()
print()

print("untuk N=10")
fig = plt.figure()
ax = fig.gca(projection="3d")

#Reset nilai U
U = 0

#Melakukan perhitungan estimasi nilai u(x,t)
len_sim = 10 #Hampiran solusi (len_sim suku pertama)
for n in range(1,len_sim):
  U = U + ((38*np.cos(n*np.pi)-18)/(n*np.pi))*np.exp(-n**2*np.pi**2*t)*np.sin(n*np.pi*x)
U = 10*x+10+U

#Visualisasi
surf = ax.plot_surface(x,t,U,
                       cmap=cm.coolwarm,
                       linewidth=0,
                       antialiased=False)
fig.colorbar(surf,
             shrink=0.5,
             aspect=5)
plt.show()
print()

print("untuk N=20")
fig = plt.figure()
ax = fig.gca(projection="3d")

#Reset nilai U
U = 0

#Melakukan perhitungan estimasi nilai u(x,t)
len_sim = 20 #Hampiran solusi (len_sim suku pertama)
for n in range(1,len_sim):
  U = U + ((38*np.cos(n*np.pi)-18)/(n*np.pi))*np.exp(-n**2*np.pi**2*t)*np.sin(n*np.pi*x)
U = 10*x+10+U

#Visualisasi
surf = ax.plot_surface(x,t,U,
                       cmap=cm.coolwarm,
                       linewidth=0,
                       antialiased=False)
fig.colorbar(surf,
             shrink=0.5,
             aspect=5)
plt.show()
print()

#c.
print('''c.
Perhatikan pada titik tengah batang (x=0.5),
suhunya adalah konstan seiring pertambahan t
''')
```
