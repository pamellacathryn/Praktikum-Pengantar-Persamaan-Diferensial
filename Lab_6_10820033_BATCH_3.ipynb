```python
#Import Library
from scipy.integrate import odeint
from sympy import *
import matplotlib.pyplot as plt
import numpy as np

'''
Nomor 1
Metode Matriks Fundamental
x' = Ax
A = [1 -1  4
     3  2 -2
     2  1 -1]
x0 = [0 2 1]
'''

print("Nomor 1")
print()

#Definisi Symbols
t, C1, C2, C3 = symbols("t C1 C2 C3")
x, y, z = symbols("x y z", cls=Function, Function=True)

A = np.array([[1,-1,4],[3,2,-1],[2,1,-1]])
x0 = np.array ([[0],[2],[1]])

#Mencari Nilai Eigen dan Vektor Eigen
eigval, eigvec = np.linalg.eig(A)
lambda1 = eigval[0]
lambda2 = eigval[1]
lambda3 = eigval[2]
print("nilai eigen 1 :",lambda1)
print("nilai eigen 2 :",lambda2)
print("nilai eigen 3 :",lambda3) #Nilai eigen dipastikan berbeda
print()
v1 = eigvec[:,0].reshape(3,1)
v2 = eigvec[:,1].reshape(3,1)
v3 = eigvec[:,2].reshape(3,1)
print("vektor eigen 1 :",v1)
print("vektor eigen 2 :",v2)
print("vektor eigen 3 :",v3)
print()

#Matriks Fundamental
psi = Matrix([[exp(lambda1*t)*float(v1[0]), exp(lambda2*t)*float(v2[0]), exp(lambda3*t)*float(v3[0])],
              [exp(lambda1*t)*float(v1[1]), exp(lambda2*t)*float(v2[1]), exp(lambda3*t)*float(v3[1])],
              [exp(lambda1*t)*float(v1[2]), exp(lambda2*t)*float(v2[2]), exp(lambda3*t)*float(v3[2])]]) #bentuk umum matriks fundamental
psi0 = psi.subs(t,0)
psi0 = np.array([[psi0[0,0], psi0[0,1], psi0[0,2]], 
                 [psi0[1,0], psi0[1,1], psi0[1,2]],
                 [psi0[2,0], psi0[2,1], psi0[2,2]]],
                dtype="float")
soln = np.matmul(np.matmul(psi,np.linalg.inv(psi0)),x0)

#Solusi PD
print("Solusi Khusus PD")
print("x(t) =",soln[0][0])
print("y(t) =",soln[1][0])
print("z(t) =",soln[2][0])
```

```python
#Import Library
from scipy.integrate import odeint
from sympy import *
import matplotlib.pyplot as plt
import numpy as np

'''
Nomor 2
Metode dsolve (tak homogen)
x' = Ax + f(t)
A = [4  2
     3 -1]
f = [-15*t*e**(-2t) -4*t*e**(-2t)]
x0 = [7,3]
'''

print("Nomor 2")
print()

t, C1, C2 = symbols("t C1 C2")
x, y = symbols("x y", cls=Function, Function=True)
eq1 = Eq(Derivative(x(t),t), 4*x(t)+2*y(t)-15*t*exp(-2*t))
eq2 = Eq(Derivative(y(t),t), 3*x(t)-1*y(t)-4*t*exp(-2*t))
soln = dsolve((eq1,eq2))
constants = solve((soln[0].subs(t,0).subs(x(0),7), soln[1].subs(t,0).subs(y(0),3)), {C1,C2})
xsoln = expand(soln[0].rhs.subs(constants))
ysoln = expand(soln[1].rhs.subs(constants))

#Solusi Umum
print("Solusi Umum")
print("Untuk x(t) ->", soln[0])
print("Untuk y(t) ->", soln[1])
print()

#Solusi Khusus
print("Solusi Khusus")
print(f"x(t) = {xsoln}")
print(f"y(t) = {ysoln}")
```

```python
#Import Library
from scipy.integrate import odeint
from sympy import *
import matplotlib.pyplot as plt
import numpy as np

'''
Bidang/Potret Fase (dsolve odeint)
x' = Ax
A = [4  2
     3 -1]
x0 = [i,i]
i = [1,2,3,4,5]
'''

print("Nomor 3")
print()

A = np.array([[4,2],[3,-1]])
eigval= np.linalg.eig(A)
lambda1 = eigval[0][0]
lambda2 = eigval[0][1]
print("nilai eigen 1 :",lambda1)
print("nilai eigen 2 :",lambda2)

#Titik Kesetimbangan
x,y = symbols('x y', real = True)
xy_crit = solve([Eq(0,4*x+2*y),
                 Eq(0,3*x-1*y)],
                [x,y])
print("titik kesetimbangan =",xy_crit)
print()

#Menggambar Bidang Fase
x = np.linspace(-10,10,20)
y = np.linspace(-10,10,20)
x,y = np.meshgrid(x,y)

plt.quiver(x,y, 4*x+2*y, 3*x-1*y)
plt.xlabel("x")
plt.ylabel("y")

#Menggambar beberapa kurva solusi
def model(U,t):
  x,y = U
  dxdt = 4*x+2*y
  dydt = 3*x-1*y
  return [dxdt,dydt]

for i in [1,2,3,4,5]:
  tspan = np.linspace(0,50,200)
  z0 = [i,i]
  ys = odeint(model, z0, tspan)
  plt.plot(ys[:,0],ys[:,1])

plt.title("Diagram Fasa")
plt.xlim([-10,10])
plt.ylim([-10,10])
plt.show()

print('''
Interpretasi:
titik kesetimbangan berada pada (0,0)
karena nilai eigennya real dan berbeda tanda, maka titik kesetimbangan (0,0) berupa saddle point
karena solusi-solusi menjauhi titik kesetimbangan, maka titik kesetimbangan (0,0) tidak stabil
''')
```

```python
#Import Library
from scipy.integrate import odeint
from sympy import *
import matplotlib.pyplot as plt
import numpy as np

'''
Nomor 4
Interaksi dua spesies
x' = x*(1-x+0.5*y)
y' = y*(2.5-1.5*y+0.25*x)
'''

print("Nomor 4")
print()

#Titik Kesetimbangan
x,y = symbols('x y', real = True)
xy_crit = solve([Eq(0,x*(1-x+0.5*y)),
                 Eq(0,y*(2.5-1.5*y+0.25*x))],
                [x,y])
print("titik kesetimbangan =",xy_crit)
print()

#Menggambar Bidang Fase
x = np.linspace(-0.5,3,20)
y = np.linspace(-0.5,3,20)
x,y = np.meshgrid(x,y)

plt.quiver(x,y,x*(1-x+0.5*y),y*(2.5-1.5*y+0.25*x))
plt.xlabel("x")
plt.ylabel("y")

#Menggambar beberapa kurva solusi
def model(U,t):
  x,y = U
  dxdt = x*(1-x+0.5*y)
  dydt = y*(2.5-1.5*y+0.25*x)
  return [dxdt,dydt]

for i in [0,0.5,1]:
  for j in [0,0.5,1]:
    tspan = np.linspace(0,50,200)
    z0 = [i,j]
    ys = odeint(model, z0, tspan)
    plt.plot(ys[:,0],ys[:,1])

plt.title("Diagram Fasa")

#plot titik kritis
plt.scatter([0,0,1,2],[0,5/3,0,2],marker="o",color="red")

plt.xlim([-0.5,3])
plt.ylim([-0.5,3])
plt.show()
print()

#Interpretasi
print('''Interpretasi:
Perhatikan arah dari vektor gradien pada bidang fasa,
terlihat bahwa semua titik kesetimbangan tidak stabil kecuali pada titik kesetimbangan (2,2),
sehingga, dapat disimpulkan bahwa kedua spesies akan tetap bertahan dan hidup berdampingan (tidak ada spesies yang punah)
''')
```
