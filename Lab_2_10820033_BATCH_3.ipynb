```python
from sympy import Function, dsolve, solve, Derivative, symbols
from sympy.abc import x
p = Function('p')
sol = dsolve(Derivative(p(x), x) - ((2*x+p(x))/(3+3*p(x)**2-x)), p(x)).rhs
constants = solve([sol.subs(x,0)-0])
solusi = sol.subs(constants)
print("solusi umum =",sol)
print("solusi khusus =",solusi)
```

```python
#Plot Grafik
x = np.linspace(-1, 1, 25)
y = np.linspace(-1, 1, 25)

def diff_eq(x, y):
  result = (2*x + 2*y) / (3 + 3*(y**2) - x)
  return result

def slope_coor(slope, domain, i_x, i_y):
  result = slope*(domain - i_x) + i_y
  return result

fig, ax = plt.subplots()
fig.set_size_inches(10, 10)

for i_x in x:
  for i_y in y:
    slope = diff_eq(i_x, i_y)
    domain = np.linspace(i_x - 0.01, i_x + 0.01, 10)
    plt.plot(domain, slope_coor(slope, domain, i_x, i_y), color='r')
```

```python
'''
f = -y**2/x**2 (tidak kontinu di x=0)
df/dy = -2*y/x**2 (tidak kontinu di x=0)
karena f dan df/dy tidak kontinu di x=0, maka untuk a=0, solusi tidak dijamin ada oleh teorema E-T
untuk a=-1 dan a=1, solusi dijamin ada dan tunggal oleh teorema E-T
'''

from sympy import Function, dsolve, solve, Derivative, symbols
from sympy.abc import x
y = Function('y')

sol = dsolve(Derivative(y(x), x) + (y(x))**2/x**2, y(x)).rhs

#untuk a=-1
a=-1
constants = solve([sol.subs(x,-a)-a])
jawaban_akhir = sol.subs(constants)
print("untuk a =",a)
print("solusi umum =",sol)
print("solusi khusus =",jawaban_akhir)
print()

#untuk a=0
a=0
constants = solve([sol.subs(x,-a)-a])
jawaban_akhir = sol.subs(constants)
print("untuk a =",a)
print("solusi umum =",sol)
print("solusi khusus =",jawaban_akhir)
print()

#untuk a=1
a=1
constants = solve([sol.subs(x,-a)-a])
jawaban_akhir = sol.subs(constants)
print("untuk a =",a)
print("solusi umum =",sol)
print("solusi khusus =",jawaban_akhir)
print()

print('''Penjelasan:
untuk nilai a yang tidak memenuhi eksistensi dan ketunggalan, python akan menampilkan solusi umumnya saja''')
```
