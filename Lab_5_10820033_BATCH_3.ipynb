```python
import numpy as np
import sympy
from sympy import *
from sympy.utilities.lambdify import lambdify
from sympy.abc import t
import matplotlib.pyplot as plt
import matplotlib.colors
from sympy.utilities.lambdify import lambdify
from sympy.abc import t

'''
METODE NILAI EIGEN MATRIKS 4x4
x'=Ax
A=[1 2 3 4
   2 1 2 3
   3 2 1 2
   4 3 2 1]
x(0)=[-2 2 -2 2]
'''


A = np.array([[1,2,3,4],[2,1,2,3],[3,2,1,2],[4,3,2,1]])

#Menghitung nilai eigen dan vektor eigen
#nilai eigen
eigval, eigvec = np.linalg.eig(A)
lambda1 = eigval[0]
lambda2 = eigval[1]
lambda3 = eigval[2]
lambda4 = eigval[3]
print("Nilai Eigen")
print("nilai eigen 1:",lambda1)
print("nilai eigen 2:",lambda2)
print("nilai eigen 3:",lambda3)
print("nilai eigen 4:",lambda4) #nilai eigen berbeda
print()

#vektor eigen
v1 = eigvec[:,0].reshape(4,1)
v2 = eigvec[:,1].reshape(4,1)
v3 = eigvec[:,2].reshape(4,1)
v4 = eigvec[:,3].reshape(4,1)
print("Vektor Eigen")
print(f"vektor eigen yang berkorespondensi dengan lambda={lambda1} adalah {v1}")
print(f"vektor eigen yang berkorespondensi dengan lambda={lambda2} adalah {v2}")
print(f"vektor eigen yang berkorespondensi dengan lambda={lambda3} adalah {v3}")
print(f"vektor eigen yang berkorespondensi dengan lambda={lambda4} adalah {v4}")
print()

#Mendefinisikan variabel
t, C1, C2, C3, C4 = symbols("t C1 C2 C3 C4")
x, y, z, h = symbols("x y z h", cls=Function, Function=True)

#Masalah nilai awal
x0 = -2
y0 = 2
z0 = -2
h0 = 2

#Solusi umum
v11 = float(v1[0])
v12 = float(v1[1])
v13 = float(v1[2])
v14 = float(v1[3])

v21 = float(v2[0])
v22 = float(v2[1])
v23 = float(v2[2])
v24 = float(v2[3])

v31 = float(v3[0])
v32 = float(v3[1])
v33 = float(v3[2])
v34 = float(v3[3])

v41 = float(v4[0])
v42 = float(v4[1])
v43 = float(v4[2])
v44 = float(v4[3])

xsol = Eq(x(t), C1*exp(lambda1*t)*v11 + C2*exp(lambda2*t)*v21 + C3*exp(lambda3*t)*v31 + C4*exp(lambda4*t)*v41)
ysol = Eq(y(t), C1*exp(lambda1*t)*v12 + C2*exp(lambda2*t)*v22 + C3*exp(lambda3*t)*v32 + C4*exp(lambda4*t)*v42)
zsol = Eq(z(t), C1*exp(lambda1*t)*v13 + C2*exp(lambda2*t)*v23 + C3*exp(lambda3*t)*v33 + C4*exp(lambda4*t)*v43)
hsol = Eq(h(t), C1*exp(lambda1*t)*v14 + C2*exp(lambda2*t)*v24 + C3*exp(lambda3*t)*v34 + C4*exp(lambda4*t)*v44)
soln = np.array([xsol,ysol,zsol,hsol])
print("Solusi Umum")
print(f"x(t) = {C1*exp(lambda1*t)*v11 + C2*exp(lambda2*t)*v21 + C3*exp(lambda3*t)*v31 + C4*exp(lambda4*t)*v41}")
print(f"y(t) = {C1*exp(lambda1*t)*v12 + C2*exp(lambda2*t)*v22 + C3*exp(lambda3*t)*v32 + C4*exp(lambda4*t)*v42}")
print(f"z(t) = {C1*exp(lambda1*t)*v13 + C2*exp(lambda2*t)*v23 + C3*exp(lambda3*t)*v33 + C4*exp(lambda4*t)*v43}")
print(f"h(t) = {C1*exp(lambda1*t)*v14 + C2*exp(lambda2*t)*v24 + C3*exp(lambda3*t)*v34 + C4*exp(lambda4*t)*v44}")
print()

#Menghitung konstanta
constants = solve((soln[0].subs(t,0).subs(x(0),x0), soln[1].subs(t,0).subs(y(0),y0), soln[2].subs(t,0).subs(z(0),z0), soln[3].subs(t,0).subs(h(0),h0)),{C1,C2,C3,C4})
xsoln = expand(xsol.rhs.subs(constants))
ysoln = expand(ysol.rhs.subs(constants))
zsoln = expand(zsol.rhs.subs(constants))
hsoln = expand(hsol.rhs.subs(constants))

#Solusi khusus
print("Solusi Khusus")
print(f"x(t) = {xsoln}")
print(f"y(t) = {ysoln}")
print(f"z(t) = {zsoln}")
print(f"p(t) = {hsoln}")
print()

#Visualisasi
xFun = lambdify(t, xsoln, 'numpy')
yFun = lambdify(t, ysoln, 'numpy')
zFun = lambdify(t, zsoln, 'numpy')
hFun = lambdify(t, hsoln, 'numpy')
tspan = np.linspace(0, 5, 100)
plt.plot(tspan, xFun(tspan))
plt.plot(tspan, yFun(tspan))
plt.plot(tspan, zFun(tspan))
plt.plot(tspan, hFun(tspan))
plt.title("Dinamika Solusi Persamaan Diferensial")
plt.xlabel("t")
plt.ylabel("Dinamika")
plt.xlim([0, 5])
plt.ylim([-10,10])
plt.legend(["x(t)","y(t)","z(t)","h(t)"])
plt.show()

print('''Dari grafik, untuk waktu yang lama t->inf,
solusi x(t), y(t), z(t), dan h(t) akan menuju ke negatif tak hingga
hal ini karena keempat solusi tersebut didominasi oleh suku dengan pangkat eksponen positif yang memiliki koefisien negatif
''')
```

```python
import numpy as np
import sympy
from sympy import *
from sympy.utilities.lambdify import lambdify
from sympy.abc import t
import matplotlib.pyplot as plt
import matplotlib.colors
from sympy.utilities.lambdify import lambdify
from sympy.abc import t

'''
METODE DSOLVE (LEBIH PRESISI)
x'=Ax
A=[2  -1
   5  -1]
x(0)=[0.1 0.1]
'''

t, C1, C2 = symbols("t C1 C2")
x, y = symbols("x y", cls=Function, Function=True)
eq1 = Eq(Derivative(x(t),t), 2*x(t)-1*y(t))
eq2 = Eq(Derivative(y(t),t), 5*x(t)-1*y(t))
soln = dsolve((eq1,eq2))
constants = solve((soln[0].subs(t,0).subs(x(0),0.1), soln[1].subs(t,0).subs(y(0),0.1)), {C1,C2})
xsoln = expand(soln[0].rhs.subs(constants))
ysoln = expand(soln[1].rhs.subs(constants))

print("a.")
#Solusi Umum
print("Solusi Umum")
print("Untuk x(t) ->", soln[0])
print("Untuk y(t) ->", soln[1])
#Solusi Khusus
print("Solusi Khusus")
print(f"x(t) = {xsoln}")
print(f"y(t) = {ysoln}")
print()

print("b.")
#Visualisasi
xFun = lambdify(t, xsoln, 'numpy')
yFun = lambdify(t, ysoln, 'numpy')
tspan = np.linspace(0, 10, 100)
plt.plot(tspan, xFun(tspan))
plt.plot(tspan, yFun(tspan))
plt.title("Dinamika Solusi Persamaan Diferensial")
plt.xlabel("t")
plt.ylabel("Dinamika")
plt.xlim([0, 10])
plt.ylim([-7, 7])
plt.legend(["x(t)","y(t)"])
plt.show()
print()

print('''c. 
Dari grafik,
kedua solusi sama-sama beresonansi karena terdapat suku sinusoidal
semakin besar waktu t, amplitudo simpangan resonansi kedua solusi akan semakin besar
''')
```

```python
import numpy as np
import sympy
from sympy import *
from sympy.utilities.lambdify import lambdify
from sympy.abc import t
import matplotlib.pyplot as plt
import matplotlib.colors
from sympy.utilities.lambdify import lambdify
from sympy.abc import t

'''
METODE DSOLVE
x'=Ax
A=[ 1  2
   -5 -2]
x(0)=[4 4]
disini, x menyatakan perasaan romeo dan y menyatakan perasaan juliet
'''

t, C1, C2 = symbols("t C1 C2")
x, y = symbols("x y", cls=Function, Function=True)
eq1 = Eq(Derivative(x(t),t), x(t)+2*y(t))
eq2 = Eq(Derivative(y(t),t), -5*x(t)-2*y(t))
soln = dsolve((eq1,eq2))
constants = solve((soln[0].subs(t,0).subs(x(0),4), soln[1].subs(t,0).subs(y(0),4)), {C1,C2})
xsoln = expand(soln[0].rhs.subs(constants))
ysoln = expand(soln[1].rhs.subs(constants))

print("a.")
print('''Masalah nilai awal yang paling sesuai adalah
x(0)=[R(0) J(0)]=[4 4]
karena pada t=0,
keduanya sudah saling menyukai dengan kuantitas 4
''')
#Solusi Khusus
print("Solusi Khusus")
print(f"x(t) = {xsoln}")
print(f"y(t) = {ysoln}")
print()

print("b.")
xFun = lambdify(t, xsoln, 'numpy')
yFun = lambdify(t, ysoln, 'numpy')

tspan = np.linspace(0, 10, 1000)

x = np.linspace(-10,10,50)
y = np.linspace(-10,10,50)
x, y = np.meshgrid(x,y)

plt.figure()
plt.quiver(x,y,x+2*y,-5*x-2*y) #Angka di matriks A
plt.plot(xFun(tspan), yFun(tspan))
plt.title("Bidang/Potret Fase")
plt.xlabel("x")
plt.ylabel("y")
plt.xlim([-10, 10])
plt.ylim([-10, 10])
plt.show()
print()

print("c.")
xFun = lambdify(t, xsoln, 'numpy')
yFun = lambdify(t, ysoln, 'numpy')
tspan = np.linspace(0, 10, 100)
plt.plot(tspan, xFun(tspan))
plt.plot(tspan, yFun(tspan))
plt.title("Dinamika Solusi Persamaan Diferensial")
plt.xlabel("t")
plt.ylabel("Dinamika")
plt.xlim([0, 10])
plt.ylim([-20, 20])
plt.legend(["x(t)","y(t)"])
plt.show()
print()

print('''d.
Dari grafik pada bagian (b) dan (c),
terlihat bahwa cinta Romeo dan Juliet beresonansi naik turun suka-benci-suka-benci,
cinta mereka akan menipis semakin lama waktu dan pada akhirnya cinta mereka habis (ke 0)
(Hal ini karena nilai eigen matriks kompleks dan bagian realnya bertanda negatif)''')
```
